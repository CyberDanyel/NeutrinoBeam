## Background
Long-baseline neutrino experiments study the nature of neutrinos, in particular *neutrino oscillations* by generating neutrino beams in accelerator facilities. The success of experiments like T2K, NO$\nu$A, and MiniBooNE relies on the correct modelling of the neutrino beam. Accurate simulations which make use of random generation techniques, also known as *Monte Carlo methods*, are an essential element of these studies.

Neutrino beams are typically produced by accelerating protons to high energies against a fixed target (e. g. graphite). The highly energetic interactions between the protons and the carbon nuclei produce secondary particles, principally *mesons*, such as *pions* $\pi$ and *kaons* $K$. These are short-lived particles which produce neutrinos in their decays; the most probable decays are two-body decays of charged pions and charged kaons:

$$
\pi^+ \rightarrow \mu^+ +\nu_\mu\,, \qquad\qquad K^+ \rightarrow \mu^+ +\nu_\mu\,.
$$

## 1. PARENT-PARTICLE REST FRAME
First, let's focus on $\nu_\mu$ production in the parent-particle rest frame.
    
Consider a generic 2-body decay $A \rightarrow B + C$ with masses $m_A$, $m_B$, $m_C$. You should have already thought through all the physics of this in the seminar.

For the particle (rest-)masses, use the approximations:
\begin{align*}
m_\pi &= 140\,\text{MeV/$c^2$} \\
m_K  &= 490\,\text{MeV/$c^2$} \\
m_\mu &= 105\,\text{MeV/$c^2$}\\
m_\nu &= 500\,\text{eV/$c^2$}
\end{align*}
(How close are these to values you could look up yourself?)

In places below, we'll use the abbreviations rf refer to the rest frame and lf for the lab frame.

   ### 1.1 Write a function which calculates the energy of the product particle $C$ in the rest frame of the parent particle. (\**)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import scipy.optimize
from matplotlib.ticker import AutoMinorLocator
#Natural units used, all energy in MeV

m_pion = 140
m_kaon = 490
m_muon = 105
m_neutrino = 500e-6

def Energy_of_C_Particle(m_a, m_b, m_c):
    Energy_c = (((m_a**2))+((m_c**2))-((m_b**2)))/(2*m_a)
    return Energy_c
```

### 1.2 Use this function to calculate the energy of the $\nu_\mu$ in the parent-particle rest frame: Enu\_rf\_pi for the pion decay and Enu\_rf\_K for the kaon decay, in units of GeV. (\**)

```python
Enu_rf_pi = (Energy_of_C_Particle(m_pion, m_muon, m_neutrino))
print('Enu_rf_pi = %.3g GeV' % (Enu_rf_pi/(10**3)))
```

```python
Enu_rf_K = (Energy_of_C_Particle(m_kaon, m_muon, m_neutrino))
print('Enu_rf_K = %.3g GeV' % (Enu_rf_K/(10**3)))
```

### 1.3 (a) Write a function to calculate the 3-momentum (again in the parent-particle rest frame) of particle $C$ as a function of polar angle $\theta$ with respect to the $z$-axis and azimuthal angle $\phi$.
##### Hint: use numpy arrays

```python
def Three_Momentum_of_C_Particle(Energy_c, m_c, phi, theta):
    
    Absolute_Momentum_c = np.sqrt(((Energy_c**2)-((m_c**2))))
    
    p_x = Absolute_Momentum_c*(np.sin(theta))*(np.cos(phi))
    p_y = Absolute_Momentum_c*(np.sin(theta))*(np.sin(phi))
    p_z = Absolute_Momentum_c*(np.cos(theta))
    
    Three_Momentum = np.array([p_x, p_y, p_z])
    return (Three_Momentum)
```

#### b) Use the functions above to calculate the 3-momentum of $\nu$ for both the $\pi$ decay and the $K$ decay in the parent-particle rest frame, for a particular value $\theta=\pi/6$ and assuming the decay happens on the plane $x = 0$.

Hint: You should find that neutrinos with $\theta=\pi/6$ have 3-momentum $z$-component $p_z=26.5$ MeV/$c$ for the $\pi$ decay and $p_z=202.4$ MeV/$c$ for the $K$ decay.

```python
pnu_rf_pi = Three_Momentum_of_C_Particle(Enu_rf_pi, m_neutrino, np.pi/2, np.pi/6)
pnu_rf_K = Three_Momentum_of_C_Particle(Enu_rf_K, m_neutrino, np.pi/2, np.pi/6)

print('3-momentum of neutrino for pion decay in pion rest frame (format is p_x, p_y, p_z) = (%.3g, %.3g, %.3g) MeV/c' % (pnu_rf_pi[0], pnu_rf_pi[1], pnu_rf_pi[2]))
print('3-momentum of neutrino for kaon decay in kaon rest frame (format is p_x, p_y, p_z) = (%.3g, %.4g, %.4g) MeV/c' % (pnu_rf_K[0], pnu_rf_K[1], pnu_rf_K[2]))
```

### 1.4 Write a function that takes as its inputs the energy and 3-momentum of a particle $C$ and returns the 4-momentum. You will use this in the next parts
##### Hint: look up the function insert for numpy arrays

```python
def Four_Momentum_Creator(Energy, Three_Momentum):
    Four_Momentum = np.insert(Three_Momentum, 0, Energy)
    return Four_Momentum
```

## 2. LAB FRAME
Now let's think about the same quantities you've just been considering, this time in the lab frame. Assume the $\pi$ and $K$ are travelling in the lab frame with energy 1 GeV in direction $+z$.
### 2.1 a) Write a function which returns the value of $\beta = v/c$ for a particle travelling at velocity $v$ in a given frame, given the mass and energy of the particle in that frame.

```python
def Beta_Finder(mass, energy):
    gamma = energy/(mass)
    beta = np.sqrt(1-(1/(gamma**2)))
    return -beta
```

#### b) Calculate the value of $\beta$ of the lab frame as seen by the parent-particle rest frame using 2.1a, for each of the two 1-GeV parent particles.
##### NOTE: Be careful with the signs!

```python
beta_pion = Beta_Finder(m_pion, 10**3)
beta_kaon = Beta_Finder(m_kaon, 10**3)
print('Beta of lab frame as seen by pion rest frame = %.3g' %(beta_pion))
print('Beta of lab frame as seen by kaon rest frame = %.3g' %(beta_kaon))
```

### 2.2 a) Write a function which allows you to boost a 4-momentum along the positive z-axis

```python
def Gamma_Finder(mass, energy):
    gamma = energy/(mass)
    return gamma
def Four_Momentum_Boost(Four_Momentum, gamma, beta):
    Lorentz_Transform = np.array([[gamma, 0, 0, -beta*gamma], [0, 1, 0, 0], [0, 0, 1, 0], [-beta*gamma, 0, 0, gamma]])
    Boosted_Four_Momentum = np.matmul(Lorentz_Transform, Four_Momentum)
    return Boosted_Four_Momentum
```

#### b) For each of the two parent particles, calculate the 4-momenta of the product neutrinos boosted into the lab frame. 
Hint: a quick check that the boost is correct is to verify that the rest mass, or the 4-momentum squared, is invariant.


```python
gamma_pion = Gamma_Finder(m_pion, 10**3)
restframe_4_momentum_nu_pi = Four_Momentum_Creator(Enu_rf_pi, pnu_rf_pi)
labframe_4_momentum_nu_pi = Four_Momentum_Boost(restframe_4_momentum_nu_pi, gamma_pion, beta_pion)
print('Lab Frame Four-Momentum of Neutrino from Pion Decay (format is Energy, pc_x, pc_y, pc_z) = (%.4g, %.4g, %.4g, %.4g) MeV' % (labframe_4_momentum_nu_pi[0], labframe_4_momentum_nu_pi[1], labframe_4_momentum_nu_pi[2], labframe_4_momentum_nu_pi[3]))
```

```python
gamma_kaon = Gamma_Finder(m_kaon, 10**3)
restframe_4_momentum_nu_kaon = Four_Momentum_Creator(Enu_rf_K, pnu_rf_K)
labframe_4_momentum_nu_kaon = Four_Momentum_Boost(restframe_4_momentum_nu_kaon, gamma_kaon, beta_kaon)
print('Lab Frame Four-Momentum of Neutrino from Kaon Decay (format is Energy, pc_x, pc_y, pc_z) = (%.4g, %.4g, %.4g, %.4g) MeV' % (labframe_4_momentum_nu_kaon[0], labframe_4_momentum_nu_kaon[1], labframe_4_momentum_nu_kaon[2], labframe_4_momentum_nu_kaon[3]))

def Invariance_Checker(Four_Momentum, Boosted_Four_Momentum, Rest_Mass):
    Rest_Energy_Squared = ((Rest_Mass**2))
    Invariant_1 = ((Four_Momentum[0])**2) - ((np.sqrt((Four_Momentum[1])**2 + (Four_Momentum[2])**2 + (Four_Momentum[3])**2))**2)
    Invariant_2 = ((Boosted_Four_Momentum[0])**2) - ((np.sqrt((Boosted_Four_Momentum[1])**2 + (Boosted_Four_Momentum[2])**2 + (Boosted_Four_Momentum[3])**2))**2)
    return Invariant_1, Invariant_2
Invariant_1, Invariant_2 = Invariance_Checker(restframe_4_momentum_nu_pi, labframe_4_momentum_nu_pi, m_neutrino)
print('Ratio of invariants in pion decay %.3f' %(Invariant_2/Invariant_1))
Invariant_3, Invariant_4 = Invariance_Checker(restframe_4_momentum_nu_kaon, labframe_4_momentum_nu_kaon, m_neutrino)
print('Ratio of invariants in kaon decay %.3f' %(Invariant_3/Invariant_4))
```

### 2.3 (\**) Use the functions you have written to make a function find4mtm\_C\_lf(mA,mB,mC,theta,phi,EA) 
which returns a four-element array of the four-momentum [$E$,$pc$] of the product particle $C$ in the lab frame, with $E$ and $pc$ both in units of MeV ($p$ is the 3-momentum in Cartesian coordinates), where the inputs are the particle masses mA, mB and mC in units of MeV/$c^2$, theta and phi, the angles $\theta$ and $\phi$ already described and EA, the parent-particle energy in the lab frame in GeV.

```python
def find4mtm_C_lf(mA,mB,mC,theta,phi,EA):
    Energy_C = Energy_of_C_Particle(mA, mB, mC)
    Three_Momentum = Three_Momentum_of_C_Particle(Energy_C, mC, phi, theta)
    Four_Momentum_Rest = Four_Momentum_Creator(Energy_C, Three_Momentum)
    Gamma = Gamma_Finder(mA, EA*(10**3))
    Beta = Beta_Finder(mA, EA*(10**3))
    Four_Momentum_Lab = Four_Momentum_Boost(Four_Momentum_Rest, Gamma, Beta)
    return Four_Momentum_Lab
```

##### NOTE: the distribution generator returns values of the momentum in units of GeV/c!

```python
#import numpy as np

def log_normal2(x, mu=0, sigma=1):
    """
    Log normal PDF, depends on parameters mu and sigma, which are analogous to the 
    mu and sigma parameters of the Gaussian distribution
    """
    if type(x)!=np.ndarray:
        if x <=0 :
            return 0
        else:
            return np.exp(- 0.5 * ((np.log(x) - mu) / sigma)**2) / (x * sigma * np.sqrt(2 * np.pi))
    else:
        x[x<=0]=1e-40 #replace zero or negative values with 
        return np.exp(- 0.5 * ((np.log(x) - mu) / sigma)**2) / (x * sigma * np.sqrt(2 * np.pi))

def log_normal_max(mu=0, sigma=1):
    """
    Returns the mode (= x max value) of the log-normal distribution
    """
    return np.exp(mu - sigma**2)

def generate_log_normal_array(x_min, x_max, mu=0, sigma=1, Narray=1):
    """
    This function yields an array of values of x in the interval [x_min; x_max] distributed according
    to a log-normal distribution as defined in log_normal(x, mu, sigma)
    The array is of length Narray
    """
    pxvals=[]
    ymax = log_normal2(log_normal_max(mu,sigma), mu, sigma)
    lenout=0
    Nchunk=Narray
    while lenout<Narray:
        px = (x_max - x_min) * np.random.rand(Nchunk) + x_min
        py = np.random.rand(Nchunk) * ymax
        lognormx=log_normal2(px, mu, sigma)   
        pxvals=np.append(pxvals,px[lognormx>py])
        lenout=len(pxvals)
    return pxvals[:Narray]
```
